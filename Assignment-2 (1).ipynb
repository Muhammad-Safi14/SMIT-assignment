# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
marks = float(input("Enter ur marks: "))
if marks >= 90:
    print("A") 
elif marks >= 80:
    print("B")
elif marks >= 70:
    print("C")
elif marks >= 60:
    print("D")
else:
    print("Fail")
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
day_num = int(input("Enter a number from 1-7:"))
if day_num == 1:
    print("Monday")
elif day_num == 2:
    print("Tuesday")
elif day_num == 3:
    print("Wednesday")
elif day_num == 4:
    print("Thursday")
elif day_num == 5:
    print("Friday")
elif day_num == 6:
    print("Saturday")
elif day_num == 7:
    print("Sunday")
else:
    print("Give the correct day.")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
mon_num = int(input("Enter the number from 1-12:"))
if mon_num == 1 or 2 or 12:
    print("The season is winter.")
elif mon_num == 3 or 4 or 5:
    print("The season is spring.")
elif mon_num == 6 or 7 or 8:
    print("The season is summer.")
else:
    print("The season is autumn.")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
num1 = float(input("Enter a number:"))
num2 = float(input("Enter a number:"))
opr = input("+,-,*,/")
if opr == "+":
    print(num1 + num2)
elif opr == "-":
    print(num1 - num2)
elif opr == "*":
    print(num1 * num2)
elif opr == "/":
    print(num1 / num2)
else:
    print("Invalid operator.")
```
